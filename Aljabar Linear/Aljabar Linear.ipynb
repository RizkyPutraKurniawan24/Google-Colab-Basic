#### **Operation on Vector and Matrix**

##### **Addition and Subtraction on Vector**

```python
V1 = np.array([[3, 5, 7]])
V2 = np.array([[2, 4, 6]])

vector_addition = V1 + V2
vector_subtraction = V1 - V2

print(vector_addition)
print(vector_subtraction)
```

##### **Addition and Subtraction on Matrix**

```python
M1 = np.array([[3, 5, 7], [9, 11, 13]])
M2 = np.array([[2, 4, 6], [8, 10, 12]])

matrix_addition = M1 + M2
matrix_subtraction = M1 - M2

print(matrix_addition)
print(matrix_subtraction)
```

#### **Transpose on Vector and Matrix**

##### **Transpose on Vector**

```python
V1 = np.array([[3, 5, 7]])
V1_transpose = np.transpose(V1)
print(V1_transpose) #convert rows to columns
```

```python
V1_untranspose = np.transpose(V1_transpose)
print(V1_untranspose) #convert columns to rows
```

##### **Transpose on Matrix**

```python
M1 = np.array([[2, 4, 6], [3, 5, 7]])
M1_transpose = np.transpose(M1)
print(M1_transpose) #convert rows to columns
```

```python
M1_untranspose = np.transpose(M1_transpose)
print(M1_untranspose) #convert columns to rows
```

#### **Multiplication and Division on Vector and Matrix**

##### **Multiplication on Vector**

```python
V1 = np.array([[3, 5, 7]])
V2 = np.array([[2, 4, 6]])

vector_multiplication = V1 * V2

print(vector_multiplication)
```

##### **Division on Vector**

```python
V1 = np.array([[3, 5, 7]])
V2 = np.array([[2, 4, 6]])

vector_division = V1 / V2

print(vector_division)
```

##### **Multiplication on Matrix**

```python
M1 = np.array([[2, 4, 6], [3, 5, 7]])
M2 = np.array([[8, 10, 12], [9, 11, 13]])

matrix_multiplication = M1 * M2

print(matrix_multiplication)
```

##### **Division on Matrix**

```python
M1 = np.array([[2, 4, 6], [3, 5, 7]])
M2 = np.array([[8, 10, 12], [9, 11, 13]])

matrix_division= M1 / M2

print(matrix_division)
```
