# **NumPy Basic**

```python
import numpy as np
```

## **Create a Vector**

```python
a = np.array([1, 2, 3, 4, 5, 6])
b = [1, 2, 3, 4, 5, 6]

print(a)
print(b)
```

```python
a = a + 1
b = b + [1]

print(a)
print(b)
```

## **Arange vector**

```python
a = np.arange(1, 10, 0.5)

print(a)
```

## **Linspace Vector**

```python
a = np.linspace(1, 10, 4)

print(a)
```

# **Array Multidimensi / Matrix**

```python
a = np.array([(1, 2, 3), (4, 5, 6)])

print(a)
```

## **Zero Matrix**

```python
a = np.zeros((5, 5))

print(a)
```

## **Ones Matrix**

```python
a = np.ones((5, 5))

print(a)
```

## **Matrix Identity**

```python
a = np.identity(5)
b = np.eye(5)

print(a)
print(b)
```
