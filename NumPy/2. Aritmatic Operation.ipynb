```python
import numpy as np
```

# **Vector**

```python
a = [1, 2, 3, 4, 5]
b = [6, 7, 8, 9, 10]

print(a)
print(b)
```

```python
anp = np.array([1, 2, 3, 4, 5])
bnp = np.array([6, 7, 8, 9, 10])

print(anp)
print(bnp)
```

## **Vector Addition**

```python
add1 = a + b
add2 = anp + bnp

print(add1)
print(add2)
```

## **Vector Subtraction**

```python
sub1 = anp - bnp

print(sub1)
```

## **Vector Multiplication**

```python
multiply1 = anp * bnp

print(multiply1)
```

## **Vector Division**

```python
divide1 = anp / bnp

print(divide1)
```

## **Vector Square**

```python
square1 = anp ** 2

print(square1)
```

# **Multidimensi Array NumPy (Matrix)**

```python
a = np.array(([1, 2, 3], [4, 5, 6]))
b = np.array(([7, 8, 9], [10, 11, 12]))

print(a)
print(b)
```

## **Matrix Addition**

```python
matrixadd1 = a + b

print(matrixadd1)
```

## **Matrix Subtraction**

```python
matrixsub1 = a - b

print(matrixsub1)
```

## **Matrix Multiplication**

```python
matrixmultiply1 = a * b

print(matrixmultiply1)
```

```python
c = np.array(([1, 2, 3], [4, 5, 6]))
d = np.ones([3, 3])

print(c)
print(d)
```

```python
c_multiplty_d = np.dot(c, d)
c_multiplty_d2 = c.dot(d)

print(c_multiplty_d)
print(c_multiplty_d2)
```

## **Matrix Division**

```python
matrixdivide1 = a / b

print(matrixdivide1)
```

## **Matrix Square**

```python
matrixsquare1 = a ** 2

print(matrixsquare1)
```
