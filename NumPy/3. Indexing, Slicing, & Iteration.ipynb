```python
import numpy as np
```

```python
a = np.arange(10)**2
a
```

## **Indexing**

```python
print('The 1st value of a is', a[0])
print('The 7th value of a is', a[6])
print('The last value of a is', a[-1])
```

## **Slicing**

```python
print('The elements 1-6 of a are', a[0:6])
print('The 4th elements to the end of a is', a[3:])
print('The elements from the beginning to 5 from a are', a[:5])
```

## **Iteration**

```python
for i in a:
    print('Value :', i)
```
