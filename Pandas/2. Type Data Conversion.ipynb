```python
import pandas as pd
```

```python
data = {'col1' : ['1', '2', '3', 'text'],
        'col2' : ['1', '2', '3', '4']}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

## **Conversion Data Type with astype() Function**

```python
df_x = df.astype({'col2' : 'int'})
df_x
```

```python
df_x.dtypes
```

## **Conversion Data Type with to_numeric Function**

```python
df.apply(pd.to_numeric, errors = 'coerce')
```
