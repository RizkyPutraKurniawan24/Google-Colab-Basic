```python
import pandas as pd
import numpy as np
```

# **Rows Selection**

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

## **Select Rows with Logical Operation**

```python
df[(df['A'] == 1) | (df['B'] == 5)]
```

## **Select Rows with isin() Function**

```python
df[~df['A'].isin([8, 5])]
```

## **Negation (~) Operation**

```python
df[~df['A'].isin([1, 3])]
```

# **Columns Selection**

```python
n_rows = 5
n_cols = 2
cols = ['Floats', 'Integers']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns = cols)
df['Floats'] = df['Floats'].astype('float')

df.index = pd.date_range(start='2025-01-01', periods=n_rows, freq='H')
df = df.reset_index()

df['Text'] = list('ABCDE')
df
```

```python
df.dtypes
```

## **Select Column Any Data Type**

### **Numbers Only**

```python
df.select_dtypes(include='number')
```

### **Floats Only**

```python
df.select_dtypes(include='float')
```

### **Integers Only**

```python
df.select_dtypes(include='int')
```

### **Text Only**

```python
df.select_dtypes(include='object')
```

### **DateTime Only**

```python
df.select_dtypes(include='datetime')
```

### **2 Combinations**

```python
df.select_dtypes(include=['number', 'datetime'])
```
